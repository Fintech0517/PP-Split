## 不添加保护直接进行攻击

```python
import sys
sys.path.append('/home/yangjirui/data/VFL/PP-Split/')
from ppsplit.attacks.label_inference.attack import NormDirect_Attack
# from target_model.data_preprocessing.dataset import NumpyDataset
from target_model.models.CriteoNet import FirstNet,SecondNet
from target_model.data_preprocessing.preprocess_criteo import preprocess_criteo
from ppsplit.fedml_core.trainer.splitLearning import SplitNN, SplitNNClient
import torch
```

```python
args = {
    'batch_size': 256,
    "hidden_dim": 16, # 输入貌似只有28个维度 线性层输入后有16个隐藏层
    'device': torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
}
```

```python
# 数据集
data_path = '/home/dengruijun/data/FinTech/DATASET/kaggle-dataset/criteo/mini_creditcard.csv'
train_loader, test_loader = preprocess_criteo(batch_size=args['batch_size'])

# 模型
model_1 = FirstNet(input_dim=31,hidden_dim=args['hidden_dim'])
model_2 = SecondNet(hidden_dim=args['hidden_dim'])
```

```python
attack = NormDirect_Attack()
attack.train_and_attack(protect_method='None',first_net=model_1,seconde_net=model_2,
                        train_loader=train_loader,test_loader=test_loader,
                        device=args['device'],)
```

## 添加保护的攻击结果

可选的攻击方法有：


`ISO`,`MAX_NORM`和`Marvell`
注意ISO有超参数可选

### 1. iso噪声

```python
# 超参数
args['t'] = 0.005

# 模型
model_1 = FirstNet(input_dim=31,hidden_dim=args['hidden_dim'])
model_2 = SecondNet(hidden_dim=args['hidden_dim'])

# 攻击
attack = NormDirect_Attack()
attack.train_and_attack(protect_method='ISO',first_net=model_1,seconde_net=model_2,
                        train_loader=train_loader,test_loader=test_loader,
                        device=args['device'],t=args['t'])
```

### 2.MAX_NORM防御

```python
# 超参数
args['t'] = 0.005

# 模型
model_1 = FirstNet(input_dim=31,hidden_dim=args['hidden_dim'])
model_2 = SecondNet(hidden_dim=args['hidden_dim'])

# 攻击
attack = NormDirect_Attack()
attack.train_and_attack(protect_method='MAX_NORM',first_net=model_1,seconde_net=model_2,
                        train_loader=train_loader,test_loader=test_loader,
                        device=args['device'])
```

### 3.Marvell防御

```python
# 模型
model_1 = FirstNet(input_dim=31,hidden_dim=args['hidden_dim'])
model_2 = SecondNet(hidden_dim=args['hidden_dim'])

# 攻击
attack = NormDirect_Attack()
attack.train_and_attack(protect_method='Marvell',first_net=model_1,seconde_net=model_2,
                        train_loader=train_loader,test_loader=test_loader,
                        device=args['device'])
```
